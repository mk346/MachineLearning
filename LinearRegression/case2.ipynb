```python
# Assign the values of X AND Y
X = FuelData[['Horse Power']] # independent variable
Y = FuelData[['Fuel Economy (MPG)']] # dependent variable
```

```python
X
```

```python
Y
```

```python
# data preparation to perform linear regression
# using train_test_split to generate train data and test data
# test size in our case is 0.3 meaning train data will be 70% while test data 30%
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3, random_state=42)
```

```python
# in our case should be 70 rows which is equvalent to 70% of 100 rows
X_train
```

```python
# test data should be equal to 30 rows which 30% of 100 rows
X_test.shape
```

```python
# should be equavalent to X_test
y_test.shape
```

```python
# should be equal to X_train
y_train.shape
```

```python
# traing the model
regressor = LinearRegression(fit_intercept=True)
regressor.fit(X_train, y_train)
```

```python
LinearRegression()
```

```python
# print the slope or the gradient of the line
print('Linear Model Coefficient (m)',regressor.coef_)
# note that the slope is negative
```

```python
# print the y intercept
print('Linear Model Coefficient (b)',regressor.intercept_)
```

```python
# forcing the y intercept to 0 using fit_intercept
regressor2 = LinearRegression(fit_intercept=False)
regressor2.fit(X_train, y_train)
```

```python
# print the slope
print('Linear Model Coefficient (m)',regressor2.coef_)
# note that the slope is positive
```

```python
# print the y intercept which should be 0
print('Linear Model Coefficient (b)',regressor2.intercept_)
```

```python
# testing the model
y_predict = regressor.predict(X_test)
```

```python
y_predict
```

```python
# plot the X_test and y_predict values for visualization
plt.scatter(X_test, y_predict,color='blue')
plt.ylabel('Fuel Economy (MPG)')
plt.xlabel('Horse Power')
plt.title('Fuel Economy vs Horse Power (Using test data vs predict dataset)')
```
